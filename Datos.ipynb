# Leemos y pre procesamos los datos

```python
variacion_porcentual = pd.read_csv('RolaMadera - Hoja 1.csv', header=None)

variacion_porcentual
```

```python
print(variacion_porcentual.shape)
variacion_porcentual_col = variacion_porcentual.values.T

print(variacion_porcentual_col)
print(variacion_porcentual_col.shape)
```

# Convertimos del formato de cada departemento en filas a cada departameno en colúmnas 

```python
chuquisaca = variacion_porcentual_col[:, 0]
la_paz = variacion_porcentual_col[:, 1]
cochabamba = variacion_porcentual_col[:, 2]
oruro = variacion_porcentual_col[:, 3]
potosi = variacion_porcentual_col[:, 4]
tarija = variacion_porcentual_col[:, 5]
santa_cruz = variacion_porcentual_col[:, 6]
beni = variacion_porcentual_col[:, 7]
pando = variacion_porcentual_col[:, 8]
```

```python
df = pd.DataFrame.from_dict({'Chuquisaca': chuquisaca, 
                             'La Paz': la_paz, 
                             'Cochabamba': cochabamba, 
                             'Oruro': oruro, 
                             'Potosí': potosi, 
                             'Tarija': tarija, 
                             'Santa Cruz': santa_cruz, 
                             'Beni': beni, 
                             'Pando': pando})

df.head()
```

## Guardamos los datos pre procesados

```python
df.to_csv('metros_cubicos_madera.csv')
```

```python
df.isna().sum()
```

## Cambiamos los valores faltantes por ceros

```python
df_num = df.replace('  - ', '0')

for col_name, row in df_num.iteritems():
    df_num[col_name] = pd.to_numeric(row, downcast='float')
    
df_num.head()
```

## Graficamos la temporalidad del volumen de la tala de árboles a lo largo de los años para cada departamento

```python
for col_name, row in df_num.iteritems():
    print(col_name)
    serie = [val for val in row if val > 0]
    years = [year for year, val in zip(range(1998, 2018), row) if val > 0]
    
    plt.ylabel('Metros Cubicos Rola')
    plt.xticks(np.arange(len(years)), years)
    plt.tick_params(axis ='x', rotation = 45) 
    plt.plot(serie)
    plt.plot(serie, 'ro')
    plt.show()
```

# Especies

```python
especies = pd.read_csv('especies - Hoja 1.csv')

especies.replace('  -   ', '0', inplace=True)
especies.iloc[:, 1:] = especies.iloc[:, 1:].astype(float)

especies.head()
```

```python
from IPython.display import display
```

```python
for i, year in enumerate(range(1998, 2018)):
    especies_list = especies.sort_values(by=[str(year)], ascending=False).iloc[:5, 0].tolist()
    counts_list = especies.sort_values(by=[str(year)], ascending=False).iloc[:5, i+1].tolist()

    print('Año:', year)
    display(pd.DataFrame.from_dict({'Especie': especies_list, 'Metros Cubicos Rola': counts_list}))
```
